```python
import pandas as pd
import os,sys
import numpy as np
```

```python
input_path = "/Users/melek/switch-2.0.9./examples/3zone_toy/inputs"
fuels_file = os.path.join(input_path, 'fuels.csv')
fuels = pd.read_csv(fuels_file)
```

```python
fuels
```

```python
# Add the column if it doesn't exist, fill with 0
fuels['f_pm25_intensity'] = fuels.get('f_pm25_intensity', 0)
pm25_map = {
    "Coal":              0.000068,
    "ResidualFuelOil":   0.000014,
    "DistillateFuelOil": 0.0000007,
    "NaturalGas":        0.00000003,
    "Uranium":           0.0,
    "BioSolid":          0.0000082,
}

# Apply it to your fuels DataFrame
fuels["f_pm25_intensity"] = fuels["fuel"].map(pm25_map)
fuels["f_pm25_intensity"] = fuels["f_pm25_intensity"] * 10**3
```

```python
fuels.to_csv(fuels_file, index=False)
```
